```python
import torch
from transformers import pipeline
pipe = pipeline(
    "text-generation",
    model="TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    torch_dtype=torch.bfloat16,
    device_map="auto"
)

def get_tags(ticket_text):
    prompt = f"""<|system|>
You are a support ticket tagger. Output the top 3 categories only.
Categories: Billing, Technical Support, Account Access, General Inquiry.</s>
<|user|>
Ticket: {ticket_text}
Tags:</s>
<|assistant|>"""

    outputs = pipe(prompt, max_new_tokens=50, do_sample=True, temperature=0.1)
    return outputs[0]["generated_text"].split("<|assistant|>")[-1]

# TEST
print("Auto-Tagging Results")
test_ticket = "I was charged twice for my premium subscription this month."
print(get_tags(test_ticket))
```
